```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

df = pd.read_csv(r"C:\Users\dell\Desktop\MBGP Mentoring\Programming\Excel Data\Automobile_data.csv")
print(df.info())
df.head()
```

```python
#1. What is the average price and average mileage for convertible cars

df_s1 = df[df["body-style"] == "convertible"]
df_sol1 = pd.DataFrame(df_s1[["price","average-mileage"]].mean().__round__(3), columns=["Average Value"]).reset_index()

df_sol1 = df_sol1.rename(columns={"index":"Variable"})

df_sol1
```

```python
#2.Sort cars by company and price in descending order
df_sol2 = df.sort_values(by=["company", "price"], ascending=False)
df_sol2
```

```python
#3.Save data for 'Toyota' separately
df_sol3 = df[df["company"] == "toyota"]
df_sol3
```

```python
# 4.Which company is producing most expensive car?
df_sol4 = df [df["price"].max() == df["price"]]
df_sol4
```

```python
#5. Name the cars on basis of maximum length, highest wheelbase, maximum horsepower
required_items = ["length", "wheel-base", "horsepower"]
df_sol5 = pd.DataFrame()
for i in required_items:
    df_temp = pd.DataFrame(df[df[i] == df[i].max()]["company"])
    df_temp["Variable"] = i
    df_temp["Value"] = df[i].max()
    df_sol5 = pd.concat([df_sol5,df_temp])

df_sol5
```

```python
#6. company wise average price of cars and order them in descending order
df_sol6 = pd.DataFrame(df.groupby(by=["company"])["price"].mean())
df_sol6 = df_sol6.sort_values(by=["price"], ascending=False).round(3)

df_sol6
```

```python
#7. create a column named: "decision", If wheel-base is less than 100 and length is less than 175 then "don't buy" else "buy"

df["decision"] = np.where(((df["wheel-base"]<100) & (df["length"]<175)), "Don't Buy", "Buy")

df
```

```python
#8. Select only cars whose decision is "buy"; find average price based on company of car

df_sol8 = df[df["decision"] == "Buy"][["company", "price", "decision"]]
df_sol8 = df_sol8.groupby(by=["company"], as_index=False)["price"].mean().round(2)
df_sol8
```

```python
#9. Save all the solutions in single Excel File.

writer = pd.ExcelWriter(r"C:\Users\dell\Desktop\MBGP Mentoring\Programming\Excel Data\Automobile Data Summary.xlsx", engine = "xlsxwriter")

df_sol1.to_excel(writer,sheet_name = "Solution 1", index = False)
df_sol2.to_excel(writer,sheet_name = "Solution 2", index = False)
df_sol3.to_excel(writer,sheet_name = "Solution 3", index = False)
df_sol4.to_excel(writer,sheet_name = "Solution 4", index = False)
df_sol5.to_excel(writer,sheet_name = "Solution 5", index = False)
df_sol6.to_excel(writer,sheet_name = "Solution 6", index = False)
df.to_excel(writer,sheet_name = "Solution 7 ", index = False)
df_sol8.to_excel(writer,sheet_name = "Solution 8", index = False)

```
